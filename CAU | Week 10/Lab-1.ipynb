```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
```

```python
img = cv2.imread("../data/images/roads.png")

plt.hist(img.ravel(), bins=50, edgecolor="black")
plt.title("Histogram of grayscale intensities")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.show()
```

```python
img = cv2.imread("../data/images/roads.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(16, 10))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap="gray")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(img.ravel(), bins=50, edgecolor="black")
plt.title("Histogram of grayscale intensities")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.show();
```

```python
def global_threshold(image, T):
    return (image >= T).astype(np.uint8)

T = 150
mask_global = global_threshold(img, T)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img, cmap="gray")
axes[0].set_title(f"Original image")
axes[0].axis("off")

axes[1].imshow(mask_global, cmap="gray")
axes[1].set_title(f"Global threshold mask (T={T})")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
masked_image = img * mask_global

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original image")
axes[0].axis("off")

axes[1].imshow(mask_global, cmap="gray")
axes[1].set_title("Binary mask")
axes[1].axis("off")

axes[2].imshow(masked_image, cmap="gray")
axes[2].set_title("Masked image")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
